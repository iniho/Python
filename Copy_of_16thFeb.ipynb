```python
def double(num):

    return num * 2

def quadruple(num):

    return double(double(num))     # Uses double twice

print(quadruple(5))                # 20
```

```python
def get_grade(score):

    if score >= 90:

        return "A"

    elif score >= 80:

        return "B"

    elif score >= 70:

        return "C"

    else:

        return "F"

print(get_grade(95))    # A

print(get_grade(75))    # C
```

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

# Test the converter

temp_c = 25
temp_f = celsius_to_fahrenheit(temp_c)

print(str(temp_c) + "°C = " + str(temp_f) + "°F")

temp_f2 = 77

temp_c2 = fahrenheit_to_celsius(temp_f2)

print(str(temp_f2) + "°F = " + str(temp_c2) + "°C")

```

```python
def calculate_subtotal(price, quantity):

    return price * quantity

def calculate_tax(amount):

    tax_rate = 0.08

    return amount * tax_rate

def calculate_shipping(subtotal):

    if subtotal >= 50:

        return 0              # Free shipping

    else:

        return 5.99

def calculate_total(price, quantity):

    subtotal = calculate_subtotal(price, quantity)

    tax = calculate_tax(subtotal)

    shipping = calculate_shipping(subtotal)

    total = subtotal + tax + shipping

    return total

# Test the shopping cart

item_price = 15.99

item_quantity = 4

final_cost = calculate_total(item_price, item_quantity)

print("Final total: $" + str(round(final_cost, 2)))

# Test with cheaper order

final_cost2 = calculate_total(10, 2)

print("Final total: $" + str(round(final_cost2, 2)))
```
