```python
temp = [28, 32, 35, 29, 31, 27, 30]
max = temp[0]
min = temp[0]
for i in temp:
  if i > max:
    max = i
  if i < min:
    min =i
print(f"Highest Temperature: {max}°C\nLowest Temperature: {min}°C")

```

```python
temperatures = [28, 32, 35, 29, 31, 27, 30]
count= 0
for temp in temperatures:
  #print(temp)
  if temp < 30:
   # print("inn" , temp)
    count += 1
    continue
print(f"Hot Days (>30°C): {count}")
```

```python
temperatures = [28, 32, 35, 40, 31, 33, 30]
lists= []
for index,temp in enumerate(temperatures):
  index += 1
  lists.append(temp)
  if temp >= 40:
      for j,value in enumerate(lists):
          if value > 30:
            j += 1
            print(f"Hot Days before alert: {j}")
            print(f"Alert! Extreme temperature 40°C detected on Day {index}")
            break



```
