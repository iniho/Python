```python
numbers = [5, 12, 8, 15, 3, 20, 7]

for num in numbers:  # 5 12 8 15

    if num == 15:

        print(f"Found number = 15: {num}")

        break

    print(f"Checking {num}...")

print("Loop finished")
```

```python
count = 0

while True:  # Infinite loop!

    count += 1 # 1 2 3 ...

    print(f"Count: {count}")

    if count >= 5:

        print("Reached 5, breaking out!")

        break
```
