```python
import json
# Python dictionary
person = {
    "name": "Alice",
    "age": 25,
    "city": "NYC",
    "skills": ["Python", "JavaScript"]
}

# Convert to JSON string
json_string = json.dumps(person)
print(json_string)
print(type(json_string))    # <class 'str'>
```

```python
json_pretty = json.dumps(person, indent=2)

print(json_pretty)
```

```python
import json

# JSON string

json_string = '{"name": "Bob", "age": 30, "active": true}'

# Convert to Python dictionary

person = json.loads(json_string)

print(person)

print(type(person))         # <class 'dict'>

print(person["name"])       # Access like dictionary
```

```python

import json

data = {

    "game": "Adventure Quest",

    "level": 15,

    "score": 8500,

    "inventory": ["sword", "shield", "potion"]

}


# Save to JSON file

with open("game_save.json", "w") as file:

    json.dump(data, file, indent=2)

print("Game saved!")
```

```python
# Load from JSON file

with open("game_save.json", "r") as file:

    loaded_data = json.load(file)

print("Game loaded!")

print("Level:", loaded_data["level"])

print("Score:", loaded_data["score"])

print("Inventory:", loaded_data["inventory"])
```

Summary

json.dumps() = dictionary to JSON string

json.loads() = JSON string to dictionary

json.dump() = dictionary to JSON file

json.load() = JSON file to dictionary


```python
import json

def save_contact(name, phone, email):
    """Save a contact to JSON file"""
    contact = {
        "name": name,
        "phone": phone,
        "email": email
    }
    with open("contacts.json", "w") as file:
        json.dump(contact, file, indent=2)
    print("Contact saved!")


def load_contact():
    """Load contact from JSON file"""
    with open("contacts.json", "r") as file:
        contact = json.load(file)
    return contact
```

```python
# Test the program
save_contact("Bob Johnson", "444-0123", "bob@email.com")

```

```python
# Load it back

loaded = load_contact()
print("\nLoaded Contact:")
print("Name:", loaded["name"])
print("Phone:", loaded["phone"])
print("Email:", loaded["email"])
```
