```python
def process_scores(students):
  averages={} #store values and key in directory
  #print(students.items())
  for name, scores in students.items():
    #print(name,scores)
    avg = round(sum(scores) / len(scores), 2)
    #print(avg)
    averages[name]=avg
  return averages



def classify_grades(averages):
  result={}

  GradeA=90
  GradeB=75
  GradeC=60
  #GradeD=50

  for name,avg in averages.items():
    if avg >= GradeA:
      grade="A"
    elif avg >= GradeB:
      grade="B"
    elif avg >= GradeC:
      grade="C"
    else:
      grade="F"
    result[name]=(avg,grade)
    #print(result)
  return result


def generate_report(classified,passing_avg=70):
  print("\n===== Student Grade Report =====")

  pass_count=0
  #report=0

  for name,(avg,grade) in classified.items():
   # print(f"in: {name,(avg,grade)}, {passing_avg}")
    status ="PASS" if avg >= passing_avg  else "FAIL"
    if status=="PASS":
      pass_count += 1
    print(f"\n{name}")
    print(f"| Average: {avg: .2f} |")
    print(f"| Grade: {grade} |")
    print(f"| Status: {status} |")

  #print("\n===== End of Report =====")
  print(f"\n Total Students : {len(classified)} \n Total Passing : {pass_count} \n Total Failing : {len(classified)-pass_count}")

  return pass_count


students={
    "Vino":[85, 90, 84],
    "Mano":[60, 65, 62],
    "Rudhu":[95, 98, 96]
}
averages=process_scores(students)
classified=classify_grades(averages)
total=generate_report(classified)
#print(total)
```
